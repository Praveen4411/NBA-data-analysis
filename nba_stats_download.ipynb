```python
import pandas as pd
import requests
import time
import json
import numpy as np
pd.set_option('display.max_columns',None)
```

```python
t_url = 'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season=2019-20&SeasonType=Regular%20Season&StatCategory=PTS'
```

```python
r = requests.get(url=t_url).json()
```

```python
column_headers = r['resultSet']['headers']
column_headers
```

```python
pd.DataFrame(r['resultSet']['rowSet'], columns = column_headers )
```

```python
temp1 = pd.DataFrame(r['resultSet']['rowSet'], columns = column_headers )
temp2 = pd.DataFrame({'Year':['2019-20' for i in range(len(temp1))],
                     'Season_type':['Regular%20Season' for i in range(len(temp1))]})
temp3 = pd.concat([temp2,temp1],axis = 1)
temp3
```

```python
del temp1,temp2,temp3
```

```python
new_columns = ['Year','Season_type'] + column_headers
```

```python
pd.DataFrame(columns = new_columns)
```

```python
df = pd.DataFrame(columns = new_columns)
seasons = ['Regular%20Season','Playoffs']
years = ['2012-13','2013-14','2014-15','2015-16','2016-17','2017-18','2018-19','2019-20','2020-21','2021-22','2022-23']
for y in years:
    for s in seasons:
        final_url = 'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season='+y+'&SeasonType='+s+'&StatCategory=PTS'
        r = requests.get(url=final_url).json()
        temp1 = pd.DataFrame(r['resultSet']['rowSet'], columns = column_headers)
        temp2 = pd.DataFrame({'Year':[y for i in range(len(temp1))],
                             'Season_type':[s for i in range(len(temp1))]})
        temp3 = pd.concat([temp2,temp1],axis = 1)
        df = pd.concat([df,temp3],axis = 0)
        print({y},{s})
        lag = np.random.uniform(low = 5,high = 40)
        print({round(lag,1)})
        time.sleep(lag)
print('FINISHED')
```

```python
df.to_excel('nba_data.xlsx',index = False)
```
